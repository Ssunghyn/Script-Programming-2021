````python
def addallodd(list:list):
    sum = 0
    for i in list:
        if i % 2 == 1:
            sum += i
    return sum
````

* length(X)를 다음과 같이 정의한다 : 10진수로 표현된 X의 자릿수   
* Strange라는 숫자를 아래과 같이 정의한다.   
    * 음이 아닌 정수 중 자릿수가 1인 모든 숫자는 Strange이다.
    * length(X) > 1 인 숫자 X 는 다음 2가지를 만족하면 Strange이다.
        * X는 length(X)로 정확히 나누어 떨어진다.
        * X / length(X) 가 strnage이다.

* 이 문제에서 숫자 하나가 주어질 때, 이 숫자가 Strange인지 판단하는 함수 isStrange(num)을 작성하시오.
* isStrange(num) 함수가 재귀적인 함수가 되도록 작성해 보세요.

````python
def isStrange(num:int):
    length = lambda x : len(str(x))

    if num<0:
        return False
    elif length(num) == 1:
            return True
    else:
        if num%length(num) == 0 and isStrange(num//length(num)): return True 
        else: return False
````

````python
num = 0
print(isStrange(num))

num = 56
print(isStrange(num))

num = 100
print(isStrange(num))

num = 7882602506492313600
print(isStrange(num))

num = 6012302549522098651
print(isStrange(num))
````

#### * 코드 설명
자릿수를 구하는 함수를 만들기 위해 lenght에 lambda 식을 사용하여 자릿수를 구합니다.    
그리고 첫번째 조건에 의하여 음인 정수는 False 값을 갖게 됩니다.    
그리고 자릿수가 1인 수는 True가 됩니다.    
만약 자릿수가 1보다 크면, 두 가지 조건을 만족해야 합니다.   
X는 length(X)로 정확히 나누어 떨어지고 X/length(X) 가 strange일 경우 True를 반환하고   
그 외의 경우에는 False를 반환합니다.

### 9. 다음 문제를 정확히 읽고 적합한 코딩을 제시하시오.

* 1번 그룹, 2번 그룹, ... n번 그룹으로 번호가 매겨진 총 n개의 그룹이 존재하고, ith 그룹에는 임의의 팀원 수 ai 명이 있다.   
하나의 버스 정류장이 있고, 버스는 오직 한 대만 운영한다. 버스는 버스 정류장에 도착하고 모든 그룹은 버스를 통해 여행하길 원한다.   
하지만, 같은 그룹에 속한 팀원들은 분리되기를 원치 않는다. 그래서 그룹의 모든 팀원이 탈 수 있을 때 에만 그 그룹은 버스에 탑승한다.   
또한, 이들은 그룹의 번호순으로 여행을 하길 원한다. 예를 들어, 3번 그룹은 1번 그룹과 2번 그룹이 여행을 한 뒤에 버스를 탈 수 있거나,    
1번 그룹과 2번 그룹이 이미 버스에 탑승한 상태에 있을 때 동시에 버스를 탈 수 있다.   
마지막으로, 버스는 빈 좌석이 없을 때 에만 여행을 떠난다.   
   
* 이 문제에서 각 그룹의 팀원 수가 리스트로 주어질 때, 가능한 모든 버스의 사이즈 x(x명의 사람들이 동시에 탈 수 있음)를 반환하는 함수 solve(a) 를 작성하시오.
* [주의] 반드시 계산 복잡도를 최대한 낮추세요. 아래 예시 3번을 실행했을 때 결과가 바로 나타날 수 있도록!!

````python
import math

def solve(a):
    setList = list(set(a))
    max = setList[-1]
    divList = []
    sumList = sum(a)
    result = []
    for i in range(1, int(math.sqrt(sumList)) + 1):
        if sumList % i == 0:
            if i < max:
                divList.append(sumList // i)

            elif sumList % i == 0:
                divList.append(i)
                if sumList / i == i:
                    divList.append(sumList // i)

    divList.sort()

    for i in divList:
        subsum = 0
        check = i
        for j in a:
            subsum += j
            if subsum == i:
                subsum = 0
            elif subsum > i:
                check = 0
                break

        if check: result.append(check)

    return result
````

````python
import time

a = [1, 2, 1, 1, 1, 2, 1, 3]
b = [1, 1, 1, 1]
c = [9999] * 10 ** 5

start = time.time()
l = solve(c)
end = time.time()
print(l)
print("소요시간 : {}".format(end - start))
````

그리고 약수를 구할 경우 가장 작게 반복할 수 있도록 리스트의 합의 제곱근까지만 반복한 후 찾은 값을 나누어 이후 값을 찾았습니다.   
이때 한 버스에 한 그룹만 탄다고 하면, 그룹 인원수의 최대 값 보다 작은 약수의 값은 버스 사이즈가 될 수 없습니다.    
큰 값들은 모두 약수 리스트에 추가 시킨 후 이를 정렬 시켰습니다.    
그리고 약수 리스트의 수 만큼 for 문을 돌리고, 부분 리스트의 합을 구하기 위한 변수와 이것이 버스사이즈가 될 수 있는 지를 판단하는 변수를 선언한 후 인수로 받은 리스트의 값을 for문으로 하나씩 꺼내서 subsum에 계속 더해줍니다. 이때 subsum이 i 값과 같아지면 이는 i 사이즈 버스에 모두 탄 경우와 같음으로 다음 그룹들이 타기 위해 새로운 버스를 가져오는 개념으로 subsum을 0으로 초기화 했습니다.   
만약 subsum이 i값보다 클 경우 i사이즈 버스에 하나의 그룹이 다 탈 수 없다는 것을 의미함으로 이는 버스 사이즈가 될 수 없습니다.   
그래서 check에 False를 넣어준 후 break 문으로 j for문을 탈출 한 후 if 문으로 check를 검사해 result 리스트에 추가해주었습니다.

### 10. 이전 Assignment 2의 마지막 문제는 웹 URL로 지정된 웹페이지를 문자열로 가져와 모든 HTML 태그를 제외한 순수 텍스트를 얻어내고 그 안에 존재하는 단어를 추출하고 단어들의 총 개수를 출력하는 프로그램을 작성하는 것이었다. 이번에는 그 마지막 숙제를 그대로 확장하여 웹 URL로 지정된 웹페이지 내 순수 텍스트 안에 존재하는 각 단어들에 대해 다음 요구사항 대로 출력하는 프로그램을 작성하시오.¶
* 요구사항 1. CSS와 Javascript 내용도 모두 제거하시오 (이전 숙제에서 이미 제외하였다면 본 요구사항 무시)
    * 즉, 아래와 같은 코드는 ```<style>``` 및 ```<script>```태그 뿐만 아니라 그 안의 내용 전체를 모두 제거    
```
<style>
h1 {
  color: blue;
}
</style>
<script>
int a = 10;
...
</script>
```   
* 요구사항 2. 순수 텍스트 안에 존재하는 단어들에 대해 string 모듈을 활용하여 모든 punctuation (구두문자)를 완벽히 제거하시오.
    * 예: ['world!', ':', '+hello+', '~python$$$', '=', 'world'] ---> ['world', 'hello', 'python', 'world']
    * 모든 punctuation을 확인하는 방법은 아래와 같다.
    ```
    import string
    print(string.punctuation)
    ```
    
* 요구사항 3. 만약 punctuation (구두문자)를 제거한 이후 공백으로 판명되는 단어들은 해당 단어 리스트에서 제거하시오.   
    * 예: ['컴퓨터공학부!', '---?', '%지향~', '...'] ---> ['컴퓨터공학부', '', '지향', ''] ---> ['컴퓨터공학부', '지향']    
    &nbsp;   
* 요구사항 4. 각 단어들의 출현빈도를 사전형태로 저장하여 출력하시오.   
    * 예: ['world', 'hello', 'python', 'world', '컴퓨터공학부', '지향'] ---> {'world': 2, 'hello': 1, 'python': 1, '컴퓨터공학부': 1, '지향': 1}    
    &nbsp;  
* 요구사항 5. 다음 5개의 사이트에 대한 단어 출현빈도 결과를 출력하시오.
    * http://cse.koreatech.ac.kr   
    * https://www.koreatech.ac.kr   
    * http://www.naver.com   
    * http://www.daum.net   
    * http://www.nytimes.com   

````python
import requests
import re
import string

'''
http://cse.koreatech.ac.kr
https://www.koreatech.ac.kr
http://www.naver.com
http://www.daum.net
http://www.nytimes.com
'''
punc = string.punctuation
req = requests.get('http://cse.koreatech.ac.kr')
source = req.text

body = re.search('<body.*</body>', source, re.I|re.S)

if (body is None) :
	exit()

body = body.group()
body = re.sub('<script.*?>.*?</script>', '', body, 0, re.I|re.S)
body = re.sub('<style.*?>.*?</style>', '', body, 0, re.I|re.S)
body = re.sub('<url.*?>.*?</url>', '', body, 0, re.I|re.S)
text = re.sub('<.+?>', ' ', body, 0, re.I|re.S)
text= re.sub('&.*;|\t|\r|\n|@.+?@', ' ', text)
text = re.sub('[›"“”…`’‘＂ㆍⓒ→]','',text)

for i in text:
	for j in punc:
		text=text.replace(j,'')

strHtml = text.split()
htmlSet = list(set(strHtml))
dict = {}
for i in htmlSet:
	dict.update({i:strHtml.count(i)})

print(dict)
````

#### * 코드 설명

처음에 html의 body 부분만 살펴보기 위해 ```<body.*</body>```부분을 찾습니다.이때 대소문자를 구분하지 않고, 줄바꿈도 포합합니다.   
만약 body가 없으면 프로그램을 종료하고, 있게된다면 body.group()을 사용하여 문자열로 만들어 줍니다.   
```
body = re.sub('<script.*?>.*?</script>', '', body, 0, re.I|re.S)
body = re.sub('<style.*?>.*?</style>', '', body, 0, re.I|re.S)
body = re.sub('<url.*?>.*?</url>', '', body, 0, re.I|re.S)
```
코드를 사용하여 ```<style>```,```<script>```,```<url>``` 태그 뿐만 아니라 그 안의 내용 전체를 모두 제거합니다.   
그리고 ```text = re.sub('<.+?>', ' ', body, 0, re.I|re.S)```를 사용하여 html 태그를 모두 지워줍니다.   
```
text= re.sub('&.*;|\t|\r|\n|@.+?@', ' ', text)
text = re.sub('["“”…`’‘＂ㆍⓒ]','',text)
```

이렇게 모두 처리한 문자열을 split()을 이용하여 띄어쓰기 단위로 나누어 리스트에 저장해줍니다.   
그리고 리스트 strHtml을 set으로 변환한 뒤 list로 다시 변환하여 중복을 제거해 줍니다.
